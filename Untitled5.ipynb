```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from sklearn.linear_model import LinearRegression

# 1. Data Load karein
file_path = '/content/Nat_Gas (1).csv'
df = pd.read_csv(file_path, parse_dates=['Dates'])

# 2. Data Visualization (Trend dekhne ke liye)
def plot_data(df):
    plt.figure(figsize=(12, 6))
    plt.plot(df['Dates'], df['Prices'], marker='o', color='b', label='Historical Prices')
    plt.title('Natural Gas Prices (2020 - 2024)')
    plt.xlabel('Date')
    plt.ylabel('Price')
    plt.grid(True)
    plt.legend()
    plt.show()

# 3. Model for Estimation & Extrapolation
# Hum sine wave (seasonality) use kar sakte hain, par abhi simple interpolation + trend rakhte hain
def estimate_price(input_date_str):
    target_date = pd.to_datetime(input_date_str)

    # Dates ko numeric mein convert karna model ke liye
    df['date_numeric'] = df['Dates'].apply(lambda x: x.timestamp())

    X = df[['date_numeric']]
    y = df['Prices']

    # Linear model trend ke liye
    model = LinearRegression()
    model.fit(X, y)

    # Prediction
    target_numeric = np.array([[target_date.timestamp()]])
    price_pred = model.predict(target_numeric)[0]

    # Seasonality adjustment (thoda desi tareeka: winter months mein price +2% kar dena)
    # Natural gas usually Dec-Jan mein mehngi hoti hai
    if target_date.month in [11, 12, 1]:
        price_pred *= 1.05  # 5% boost for winter
    elif target_date.month in [6, 7, 8]:
        price_pred *= 0.95  # 5% dip for summer

    return round(price_pred, 2)

# --- EXECUTION ---
plot_data(df)

# Test: Kisi bhi date ka price check karein (Past ya Future)
test_date = '2025-01-15' # Extrapolation for next year
print(f"The estimated price for {test_date} is: ${estimate_price(test_date)}")
```
